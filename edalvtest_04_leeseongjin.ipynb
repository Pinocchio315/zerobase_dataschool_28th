```python
# 1 - 3

import pycountry

# 'code' ì»¬ëŸ¼ ì¶”ê°€
df_target['code'] = ''

# êµ­ê°€ ì½”ë“œ ì…ë ¥
for idx, row in df_target.iterrows():
    country_name = row['country']
    try:
        # ì¼ë°˜ ê²€ìƒ‰ ì‹œë„
        country = pycountry.countries.get(name=country_name)
        if country:  # Check if country is found
            df_target.loc[idx, 'code'] = country.alpha_2
        else:
            # ì¼ë°˜ ê²€ìƒ‰ ì‹¤íŒ¨ ì‹œ fuzzy ê²€ìƒ‰ ì‹œë„
            country_list = pycountry.countries.search_fuzzy(country_name)
            df_target.loc[idx, 'code'] = country_list[0].alpha_2
    except LookupError:
        print(f"Country not found: {country_name}")  # ì°¾ì§€ ëª»í•œ êµ­ê°€
        if country_name == 'Turkey':
            df_target.loc[idx, 'code'] = 'TR' # ë¬´ìŠ¨ ì´ìœ ì—ì„œì¸ì§€ í„°í‚¤ë¥¼ ì°¾ì§€ ëª»í•¨.
        else:
            df_target.loc[idx, 'code'] = 'N/A'  # Or handle the missing code in another way

check_01_03(df_target)

df_target
```
