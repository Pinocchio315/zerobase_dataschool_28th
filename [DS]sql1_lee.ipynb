- 이디야 매장 주요 분포 지역 (매장수가 많은 상위 5개 구이름, 매장 개수 출력)
- 구별 브랜드 각각의 매장 개수 조회 (구이름, 브랜드이름, 매장 개수 출력)
- 구별 브랜드 각각의 매장 개수 조회 (구이름, 스타벅스 매장 개수, 이디야 매장 개수 출력)

```python
# 스타벅스 매장 주요 분포 지역 (매장수가 많은 상위 5개 구이름, 매장 개수 출력)
query = """
    SELECT gu_name, COUNT(name) as store_count 
    FROM COFFEE_STORE 
    WHERE brand = 1 
    GROUP BY gu_name 
    ORDER BY store_count DESC 
    LIMIT 5;
    """
pin_cur.execute(query)
for item in pin_cur:
    print(item)
```

```python
# 이디야 매장 주요 분포 지역 (매장수가 많은 상위 5개 구이름, 매장 개수 출력)
query = """
    SELECT gu_name, COUNT(name) as store_count 
    FROM COFFEE_STORE 
    WHERE brand = 2 
    GROUP BY gu_name 
    ORDER BY store_count DESC 
    LIMIT 5;
    """
pin_cur.execute(query)
for item in pin_cur:
    print(item)
```

```python
# 구별 브랜드 각각의 매장 개수 조회 (구이름, 브랜드이름, 매장 개수 출력)
query = """
    select s.gu_name, b.name, count(s.name) as store_count 
    from COFFEE_STORE s, COFFEE_BRAND b  
    where s.brand = b.id 
    group by s.gu_name, s.brand 
    order by s.gu_name, b.name;
    """
pin_cur.execute(query)
for item in pin_cur:
    print(item)
```

```python
# 구별 브랜드 각각의 매장 개수 조회 (구이름, 스타벅스 매장 개수, 이디야 매장 개수 출력)
query = '''
select sbuck.gu_name, sbuck.count STARBUCKS, ediya.count EDIYA 
from 
    (select gu_name, count(brand) count from COFFEE_STORE WHERE brand = 1 GROUP BY gu_name) sbuck, 
    (select gu_name, count(brand) count from COFFEE_STORE WHERE brand = 2 GROUP BY gu_name) ediya 
where sbuck.gu_name = ediya.gu_name;
'''
pin_cur.execute(query)
for item in pin_cur:
    print(item)
```

시각화 프로젝트를 위하여 다음의 규칙으로 쿼리하여 CSV 파일로 저장합니다. (Python 코드로 작업)

- 전체 데이터를 가져오는데, 각 스타벅스 매장별로 이디야 전체 매장정보가 매칭되어 있어야 합니다. (정렬 : s_id, e_id 순)
- 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
- 데이터 프레임 출력을 해주세요. 데이터 프레임 미출력시 감점입니다.

```python
query = '''
select * from 
    (select s.id s_id, b.name s_brand, s.name s_name, s.gu_name s_gu, s.address s_address, s.lat s_lat, s.lng s_lng 
    from COFFEE_STORE s, COFFEE_BRAND b 
    where b.id = s.brand and b.name like 'STARBUCKS') sbuck, 
    (select s.id e_id, b.name e_brand, s.name e_name, s.gu_name e_gu, s.address e_address, s.lat e_lat, s.lng e_lng 
    from COFFEE_STORE s, COFFEE_BRAND b 
    where b.id = s.brand and b.name like 'EDIYA') ediya 
order by sbuck.s_id, ediya.e_id;
'''
pin_cur.execute(query)
df = pd.DataFrame(pin_cur)
df.head()
```
