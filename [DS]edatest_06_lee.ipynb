- 완료 후 결과 dataframe 변수(df_target)를 check_02_01 함수에 입력하여 채점하세요.

````python
# 2 - 1

import numpy as np

df_target['여자비율'] = np.nan

for index, row in df_target.iterrows():
  if not pd.isnull(row['여자']):
    try:
      x = round(((row['전체'] - row['남자']) / (row['여자']-row['남자'])) * 100, 6)
      df_target.loc[index, '여자비율'] = x
    except ZeroDivisionError:
      df_target.loc[index, '여자비율'] = np.nan

check_02_01(df_target)
df_target.head()
````
